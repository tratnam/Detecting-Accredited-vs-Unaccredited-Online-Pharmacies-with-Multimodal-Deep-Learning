## Web Scrape for Non-Accredited Pharmacies

```python
import pandas as pd
import shutil
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pdb
from urllib.request import urlopen, Request
import requests
```

```python
#load csv file of non-accredited pharmacy urls
df = pd.read_csv('fake-pharmacies.csv', header=None)

#convert rows to strings, remove spaces, rename column
df['url'] = df[0].str.strip()
#drop old column
df = df.drop([0], axis=1)

# strip https:// bc they only occur for some of the urls
df['url'] = df['url'].str.strip('https://')
#make sure entire strings are lowercase
df['url'] = df['url'].str.lower()

# strip 'www.' bc they only occur for some urls
#df['url'] = df['url'].str.strip('www.')

# add the https:// part back in for all of them
df['url'] = 'https://' + df['url'].astype(str)
df.head(20)
```

```python
#get a list of all the raw links
raw_links = df['url'].tolist()
raw_links
```

```python
#remove clearly problematic strings: 'https://euphoria healthcare pvt ltd.'
#'https://icenetworks ltd./www.mega-pillspharmacy.com'
# initializing K
K = 'https://euphoria healthcare pvt ltd.'
L = 'https://icenetworks ltd./www.mega-pillspharmacy.com'
 
# using list comprehension to Remove K String from String List
raw_links = [i for i in raw_links if i != K]
raw_links = [i for i in raw_links if i != L]
```

```python
#build a list of urls that can be accessed without error and scrapped 
#build a url list of the bad urls 
def sort_urls(raw_links_list):
    url_list = []
    bad_urls = []
    for each in raw_links_list:
        try:
            headers = {'user-agent': 'ds6050 (vkb6bn@virginia.edu)'}
            results = requests.get(each, headers=headers, timeout=2.0)
        except requests.exceptions.RequestException as e:
            #print('Bad URL: ' + each)
            bad_urls.append(each)
            continue
        #print(results)
        #print(each) 
        if results.status_code == 200:
            url_list.append(each)
    return url_list, bad_urls

sort_urls(raw_links)
```

### Loop through url_list to scrape text & images

```python
def url_to_df(url_list):
    for url in url_list:
        try:
            html = urlopen(url)
            headers = {'user-agent': 'Mozilla/5.0'}
            page = requests.get(url, headers)
            soup = BeautifulSoup(page.content,'html.parser')
            body = soup.body
            header = soup.header
            footer = soup.footer
        
            head = []
            for h in header.strings:
                head.append(h)
        
            foot = []
            for f in footer.strings:
                foot.append(f)
        
            text = []
            for s in body.strings:
                text.append(s)
        
            images = []    
            for item in soup.find_all('img'):
                images.append(url + item['src'])
    
            image2 = []
            for item in soup.find_all('img'):
                image2.append(item['src'])
        
            url_df = pd.DataFrame(data= {'url':[url],'text':[text], 'header':[head],'footer':[foot],
                                         'image_url':[images], 'image_source':[image2]})
            return url_df
        except: 
            pass
```
